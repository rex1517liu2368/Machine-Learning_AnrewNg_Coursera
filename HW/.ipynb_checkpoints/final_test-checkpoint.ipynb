#### Linear regression with one variable

```python
#read the dataset
ex1=np.loadtxt("machine-learning-ex1/ex1/ex1data1.txt",delimiter=',')
X=np.c_[np.ones(ex1.shape[0]),ex1[:,0]] #np.ones --> 
y=np.c_[ex1[:,1]]
print(X.shape)
```

```python
%matplotlib inline
plt.scatter(X[:,1], y, 10, c='r', marker='x', linewidths=1)
plt.xlim(4,24)
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s');
```

#### Gradient Decent

```python
def costFunction(x,y,theta=[[0],[0]]):
    m=y.size
    J=0
    h=np.dot(X,theta)
    J=1/(2*m)*np.sum(np.square(h-y))
    
    return(J)
```

```python
costFunction(X,y)
```

```python
def gradientDecent(X,y,theta=[[0],[0]],alpha=0.01,iterations=1500):
    m=y.size
    cost_list=np.zeros(iterations)
    
    for ite in np.arange(iterations):
        h=np.dot(X,theta)
        theta=theta-alpha*(1/m)*(X.T.dot(h-y))
        cost_list[ite]=costFunction(X,y,theta)
    return(theta, cost_list)
```

```python
theta,cost_list=gradientDecent(X,y)
print('theta:',theta.ravel())

plt.plot(cost_list)
plt.xlim(0,1500)
plt.ylabel('Cost J')
plt.xlabel('Iterations');
```

```python
xx=np.arange(5,23)
yy=theta[0]+theta[1]*xx

#plot gradient descent
plt.figure(figsize=(10,6))
plt.scatter(X[:,1],y,s=30,c="r",marker="x",linewidths=1)
plt.plot(xx,yy,label="Linear regression (Gradient descent)")

#Compare with Scikit-learn Linear regression
regr=LinearRegression()
regr.fit(X[:,1].reshape(-1,1),y.ravel())
plt.plot(xx,regr.intercept_+regr.coef_*xx,label='Linear regression(Scikit-learn GLM)')

plt.xlim(4,24)
plt.xlabel("Population of City in 10,000s")
plt.ylabel("Profit in $10,000s")
plt.legend(loc=4);
```

```python
B0=np.linspace(-10,10,50)
B1=np.linspace(-1,4,50)
xx,yy=np.meshgrid(B0,B1,indexing="xy")
Z=np.zeros((B0.size,B1.size))

#Calculate Z-values(cost) based on grid of coefficients
for (i,j), v in np.ndenumerate(Z):
    Z[i,j]=costFunction(X,y,theta=[[xx[i,j]],[yy[i,j]]])
    
fig=plt.figure(figsize=(15,5))

#Left Plot
ax1=fig.add_subplot(121)
CS=ax1.contour(xx,yy,Z,np.logspace(-2,3,20),cmap=plt.cm.jet)
ax1.scatter(theta[0],theta[1],c='r')

#Right plot
ax2=fig.add_subplot(122,projection='3d')
ax2.plot_surface(xx,yy,Z,rstride=1,cstride=1,alpha=0.5,cmap=plt.cm.jet)
ax2.set_zlabel("Cost",fontsize=14)
ax2.set_zlim(Z.min(),Z.max())
ax2.view_init(elev=15,azim=230)

#setting common to both plots
for ax in fig.axes:
    ax.set_xlabel(r'$\theta_0$',fontsize=17)
    ax.set_ylabel(r'$\theta_1$',fontsize=17)
```
